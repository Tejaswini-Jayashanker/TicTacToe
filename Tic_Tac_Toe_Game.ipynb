```python
import numpy as np
```

```python
class Board(object):
    def __init__(self):
        self.grid = np.zeros((3,3))
        rows,columns = np.where(self.grid == 0)
        self.avail_moves = list(zip(rows,columns))
        self.symbols = {0: "", 1: "X", -1 : "O"}
        self.status = "No"

    def print_board(self):
        print(f"""
{self.symbols[self.grid[0,0]]}|{self.symbols[self.grid[0,1]]}|{self.symbols[self.grid[0,2]]}
-----
{self.symbols[self.grid[1,0]]}|{self.symbols[self.grid[1,1]]}|{self.symbols[self.grid[1,2]]}
-----
{self.symbols[self.grid[2,0]]}|{self.symbols[self.grid[2,1]]}|{self.symbols[self.grid[2,2]]}
        """)

    def update_amoves(self):
        rows,columns = np.where(self.grid == 0)
        self.avail_moves = list(zip(rows,columns))
        
    def update_grid(self,row,col,symbol):
        self.update_amoves()
        if (row,col) in self.avail_moves:
            self.grid[row,col] = symbol
            return 1
        else:
            print("Wrong Move")
            return 0
    def game_status(self):
        self.update_amoves()
        if ( ( abs(np.sum( self.grid[0] )) == 3 ) or ( abs(np.sum(self.grid[1])) == 3 ) or ( abs(np.sum(self.grid[2])) == 3 ) ):
            self.status = "Finish"
        elif( ( abs(np.sum(self.grid[:0])) == 3 ) or ( abs(np.sum(self.grid[:,1])) == 3 ) or ( abs(np.sum(self.grid[:,2])) == 3 ) ):
            self.status = "Finish"
        elif( (abs(self.grid[0,0] + self.grid[1,1] + self.grid[2,2]) == 3) or (abs(self.grid[0,2] + self.grid[1,1] + self.grid[2,2]) == 3) ):
            self.status = "Finish"
        elif self.avail_moves == None:
            self.status = "Draw"
```

```python
class Player(object):
    def __init__(self):
        self.symbol = {1: 1, 2: -1}
        
    def move(self, player_number, board):
        while (True):
            print(f"Player {player_number} turn")
            print(f"Available moves: {board.avail_moves}")
            coord = input("Index number")
            row,col = coord.split(",")
            if board.update_grid(int(row),int(col),self.symbol[player_number]):
                break
```

```python
#if __name__ == "__main__":
board = Board()
board.print_board()
p = Player()
player = 1
while board.status == "No":
    p.move(player,board)
    board.print_board()
    board.game_status()
    if board.status == "Draw" or board.status == "Finish":
        break
    player = 2 if player == 1 else 1
if board.status == "Finish":
    print(f"Player {player} has won!!!")
elif board.status == "Draw":
        print("GAME DRAW!")
```
